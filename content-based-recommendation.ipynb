## Data Preparation

```python
# Total ratings for books by users
total_ratings_df = pd.DataFrame(ratings_df.
                                groupby("ISBN")["Book-Rating"].
                                count())

total_ratings_df.rename(columns={"Book-Rating": "Total-Ratings"}, 
                        inplace = True)

total_ratings_df.sort_values(by=["Total-Ratings"], ascending=False).head()
```

```python
total_ratings_df["Average-Rating"] = pd.DataFrame(ratings_df.
                                                  groupby("ISBN")["Book-Rating"].
                                                  mean())

total_ratings_df.reset_index(inplace = True)
```

```python
total_ratings_df.head()
```

```python
final_df = ratings_df.merge(total_ratings_df, on="ISBN")

final_df.head()
```

```python
# Total number of Book a user has rated
user_count_isbn = pd.DataFrame(ratings_df.
                               groupby("User-ID")["ISBN"].
                               count())

user_count_isbn.rename(columns={"ISBN": "Books-User-Rated"}, 
                       inplace=True)

user_count_isbn.reset_index(inplace = True)
```

```python
user_count_isbn.sort_values(by=["Books-User-Rated"], ascending=False).head()
```

```python
user_count_isbn["Books-User-Rated"].describe()
```

It is evident that more than 50% of users have only rated once (1 book only).

```python
preprocessed_df = final_df.merge(user_count_isbn, on = "User-ID")

preprocessed_df.head()
```

```python
preprocessed_df.shape
```

```python
filtered_df = preprocessed_df[preprocessed_df["Total-Ratings"] > 200]
filtered_df.shape
```

```python
filtered_df = filtered_df[filtered_df["Books-User-Rated"] > 10]
filtered_df.shape
```

## Pivot Table


```python
books_matrix = filtered_df.pivot_table(index="User-ID", 
                                       columns="ISBN", 
                                       values="Book-Rating")

books_matrix.shape
```

```python
books_matrix.head()
```

## Recommendations

```python
def get_recommendations(books_matrix, filtered_df, ISBN, NUM_BOOKS = 11):
    
    selected_isbn = books_matrix[ISBN]
    similar_to_isbn = books_matrix.corrwith(selected_isbn)
    
    correlation = pd.DataFrame(similar_to_isbn, columns=["Pearson_coeff"])
    correlation.dropna(inplace=True)
    correlation.reset_index(inplace=True)
    
    summary = correlation.merge(total_ratings_df, on="ISBN")
    summary = summary[summary["Total-Ratings"] > 200].sort_values(by=["Pearson_coeff"], ascending=False)
    
    return summary.head(NUM_BOOKS)


def get_booksName(recommendations_df, books_df):
    
    summary = recommendations_df.merge(books_df, on="ISBN")
    return summary
```

```python
get_booksName(get_recommendations(books_matrix, filtered_df, "0060938455", NUM_BOOKS = 11), 
              books_df)
```
